Python es un lenguaje de programacin interpretado de cdigo abierto, multiparadigma, aunque principalmente orientado a objetos de alto nivel.

- **Interpretado**: Ejecuta el cdigo directamente, sin necesidad de compilacin.


- **Cdigo abierto**: Tambin llamado **"open source"**, se refiere a un lenguaje que se puede distribuir libremente y puede ser usado sin ninguna restriccin.


- **Multiparadigma**: Soporta varios paradigmas de programacin como: orientacin a objetos, estructurada, programacin imperativa y programacin funcional (aunque principalmente la orentacion a objetos).


- **Alto nivel**: Son aquellos que se encuentran mas cercanos al lenguaje natural (al lenguaje que usamos para comunicarnos) que al lenguaje maquina.

Python fue creado por **Guido van Rossum**, un programador holands a finales de los 80, y se caracteriza principalmente por ser un lenguaje de programacin fcil y sencillo de aprender. Su cdigo es ms organizado, evitando el uso de signos de punto y coma al final de cada sentencia y tambin evita el uso de llaves para especificar bloques de cdigo.

- Portabilidad: Python es compatible con todos los sistemas operativos.
- Comunidad: Python cuenta con una gran comunidad en todas las diferentes areas en la que es utilizado.
