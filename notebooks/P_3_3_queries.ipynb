```python
#funci√≥n ejecutar para queries
#actualizar user/password antes de ejecutar

def select_from_table(query, database, host = "localhost", user = "root", password = "Hackaboss_2024"):

    db = mysql.connector.connect(host     = "localhost",
                                 user     = "root",
                                 password = "Hackaboss_2024",
                                 database = "yfinance_stocks")

    cursor = db.cursor()

    cursor.execute(query) #ejecutar la query
    column_names = cursor.column_names #columnas de la tabla

    #guardar los datos de la tabla
    data = cursor.fetchall()

    cursor.close()
    db.close()
    
    return data, column_names
```

```python
#query global de stocks
table_name = "stocks"

query = f"""SELECT * FROM {table_name};"""

database = "yfinance_stocks"

data, column_names = select_from_table(query = query, database = database)

pd.DataFrame(data = data, columns = column_names)
```

```python
#precio promedio de cierre ajustado (adj_close) de las acciones de NVDA en 2023

table_name = "historical_prices"

query = f"""
SELECT AVG(adj_close) AS avg_adj_close
FROM historical_prices
WHERE ticker = "NVDA" 
AND date BETWEEN "2023-01-01" AND "2023-12-31";
"""

database = "yfinance_stocks"

data, column_names = select_from_table(query = query, database = database)

pd.DataFrame(data = data, columns = column_names)
```

```python
#comparaci√≥n de capitalizaci√≥n de mercado (market_cap) de NVDA con las dem√°s empresas de su misma industria

table_name = "stocks"

query = f"""
SELECT industry, name, market_cap, ticker
FROM stocks
WHERE industry = (
    SELECT industry 
    FROM stocks 
    WHERE ticker = "NVDA"
)
ORDER BY market_cap DESC;
"""

database = "yfinance_stocks"

data, column_names = select_from_table(query = query, database = database)

pd.DataFrame(data = data, columns = column_names)
```

```python
#las 5 fechas con mayor volumen de transacciones

table_name = "cashflow"

query = f"""
SELECT date, volume
FROM historical_prices
WHERE ticker = "NVDA"
ORDER BY volume DESC
LIMIT 5;
"""

database = "yfinance_stocks"

data, column_names = select_from_table(query = query, database = database)

pd.DataFrame(data = data, columns = column_names)
```

```python
#flujo de caja operativo (operating_cash_flow) y el gasto de capital (capital_expenditure) en 2023

table_name = "cashflow"

query = f"""
SELECT date, operating_cash_flow, capital_expenditure
FROM cashflow
WHERE ticker = "NVDA"
AND date BETWEEN "2022-01-01" AND "2023-12-31";
"""

database = "yfinance_stocks"

data, column_names = select_from_table(query = query, database = database)

pd.DataFrame(data = data, columns = column_names)
```

```python
#crecimiento del ingreso neto (net_income) por a√±o

table_name = "income_statement"

query = f"""
SELECT YEAR(date) AS year, SUM(net_income) AS total_net_income
FROM income_statement
WHERE ticker = "NVDA"
GROUP BY YEAR(date)
ORDER BY year;
"""

database = "yfinance_stocks"

data, column_names = select_from_table(query = query, database = database)

pd.DataFrame(data = data, columns = column_names)
```

```python
#Relaci√≥n de las deudas netas (net_debt) con los activos totales (total_assets) a lo largo del tiempo

table_name = "balance_sheet"

query = f"""
SELECT date, net_debt, total_assets, (net_debt / total_assets) * 100 AS debt_to_assets_percentage
FROM balance_sheet
WHERE ticker = "NVDA"
ORDER BY date;
"""

database = "yfinance_stocks"

data, column_names = select_from_table(query = query, database = database)

pd.DataFrame(data = data, columns = column_names)
```

```python
#principales recomendaciones de analistas

table_name = "upgrades_downgrades"

query = f"""
SELECT date, firm, action, to_grade, from_grade
FROM upgrades_downgrades
WHERE ticker = "NVDA"
ORDER BY date DESC
LIMIT 10;
"""

database = "yfinance_stocks"

data, column_names = select_from_table(query = query, database = database)

pd.DataFrame(data = data, columns = column_names)
```

```python
#dividendos pagados por NVDA por a√±o

table_name = "upgrades_downgrades"

query = f"""
SELECT YEAR(date) AS year, SUM(dividends) AS total_dividends
FROM dividends
WHERE ticker = "NVDA"
GROUP BY YEAR(date)
ORDER BY year;
"""

database = "yfinance_stocks"

data, column_names = select_from_table(query = query, database = database)

pd.DataFrame(data = data, columns = column_names)
```
